# Map

Uma função é aplicada a cada elemento de uma lista resultando em uma nova lista de mesmo comprimento que a original.

```python
# Lista original
frutas = ["laranja", "abacaxi", "pêssego", "morango"]

# Função que será usada para a operação de 'map'
def para_maiusculas(palavra):
    return palavra.upper()
```

aplicando a função a cada elemento da lista usando um loop `for` (para cada)

```python
for palavra in frutas:
    fruta_maiuscula = para_maiusculas(palavra)
    print(fruta_maiuscula)
```

mas isso não nos retorna uma lista!

```python
def frutas_em_maiusculas(lista_de_frutas):
    for fruta in lista_de_frutas:
        fruta_maiuscula = para_maiusculas(fruta)
        print(fruta_maiuscula)
```

estamos apenas aplicando a função a cada elemento da lista mas nao estamos criando uma lista nova

```python
frutas_em_maiusculas(frutas)
```

além disso, a função `frutas_em_maiusculas()` tem um efeito colateral: ela não
apenas processa os dados da lista como também escreve na tela (o efeito colateral)

```python
# Esta função vai fazer um 'map' propriamente
# porque ela vai criar uma NOVA lista mapeando
# os dados da lista de entrada

def frutas_maiusculas(lista_de_frutas):
    nova_lista = list()
    for fruta in lista_de_frutas:
        fruta_maiuscula = para_maiusculas(fruta)
        nova_lista.append(fruta_maiuscula)
    
    return nova_lista
```

```python
frutas_maiusculas(frutas)
```

```python
for fruta in frutas_maiusculas(frutas):
    print(fruta)
```

## map()

```python
# Lista original
frutas = ["laranja", "abacaxi", "pêssego", "morango"]

# Função que será usada para a operação de 'map'
def para_maiusculas(palavra):
    return palavra.upper()
```

```python
list(map(para_maiusculas, frutas))
```

```python

for fruta_maiuscula in map(para_maiusculas, frutas):
    print(fruta_maiuscula)
```

```python
em_maiusculas = list(map(para_maiusculas, frutas))
```

### Exemplo

Map com a função `f(x) = x + 2`

```python
valores =  [1,2,3,4,5]

def f(x):
    return x + 2

list(map(f, valores))
```

Função dobro(x) = X * 2

```python
def dobro(x):
    return x * 2

list(map(dobro, valores))
```
